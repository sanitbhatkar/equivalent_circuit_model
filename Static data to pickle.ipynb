__Function for File Storage to Dictionary__

```python
# file_name = 'A123_OCV_N05_S1.csv'

def data_dict(file_name,use_cols,col_name):
    
    # Reading the CSV file
    
    df = pd.read_csv(file_name, usecols = use_cols)
    df.columns = col_name


    # Dictionary used for saving the data

    return df.to_dict('series')
```

__Nested key printing function__

```python
def nest_key(d, indent=0):
    
    for key, value in d.items():
        
        print('\t' * indent + f'{key}')
        
        if isinstance(value, dict):
            
            nest_key(value, indent+1)
```
